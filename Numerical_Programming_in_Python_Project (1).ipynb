## ***1. Know Your Data***

### Import Libraries

```python
# Import Libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Dataset Loading

```python
df = pd.read_csv('Airbnb NYC.csv', encoding= 'unicode_escape')
```

```python
# Dataset Rows & Columns count
df.shape
```

# **Duplicate Values**

```python
df.duplicated().sum()
```

```python
#checkin the empty/null values
df.isnull().sum()
```

```python
# Treating null values
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
mean_date = df['last_review'].mean()
df['last_review'].fillna(mean_date, inplace=True)
df['reviews_per_month'].fillna(0, inplace=True)
df['name'].fillna('Unnamed Listing', inplace=True)
df['host_name'].fillna('Unknown Host', inplace=True)
df.info()
```

```python
df.isnull().sum()
```

So there are no null values.

## ***4. Data Vizualization, Storytelling & Experimenting with charts***

#### Chart - 1

**Distribution of Airbnb Prices**

```python
plt.figure(figsize=(10,5))
sns.histplot(df['price'], bins=50, kde=True)  # Reduced bins for clarity
plt.xlim(0, 500)
plt.title('Distribution of Airbnb Prices')
plt.xlabel('Price ($)')
plt.ylabel('Count of Listings')
plt.show()
```

##### 1. Why did you pick the specific chart?

I chose the histogram because it’s the most effective way to visualize the distribution of a continuous variable — in this case, Airbnb prices. It helps reveal how prices are spread across listings, highlights the most common pricing ranges, and clearly shows skewness or the presence of outliers. This insight is crucial for identifying typical pricing behavior and guiding strategic decisions, such as setting competitive rates or filtering out extreme values.

##### 2. What is/are the insight(s) found from the chart?

The histogram reveals that most Airbnb listings are priced between $50 and $200, with a sharp concentration around the $100 mark, indicating it as the most common or competitive price point. The distribution is right-skewed, showing a long tail of high-priced listings, which are relatively rare.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, the insights from the chart can significantly impact business decisions. By identifying the most common pricing range, Airbnb hosts and the platform itself can optimize pricing strategies to stay competitive and attract more bookings.

#### Chart - 7

**Top 10 Hosts with Most Listings**

```python
# Chart - 7 visualization code
plt.figure(figsize=(14, 6))
top_hosts = df['host_name'].value_counts().nlargest(10)
sns.barplot(x=top_hosts.index, y=top_hosts.values, palette="viridis")

plt.title('Top 10 Hosts with Most Listings', fontsize=18, fontweight='bold')
plt.xlabel('Host Name', fontsize=14)
plt.ylabel('Number of Listings', fontsize=14)
plt.xticks(rotation=30, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

##### 1. Why did you pick the specific chart?

This bar chart clearly highlights the top 10 hosts with the most listings, making it easy to identify major contributors.

##### 2. What is/are the insight(s) found from the chart?

A small number of hosts, like Michael and David, control a disproportionately large number of listings.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, it helps identify key partners for promotions, but over-reliance on a few hosts may reduce platform diversity and hurt long-term customer trust.

#### Chart - 8

**Average Price per Neighbourhood Group**

```python
# Chart - 8 visualization code
plt.figure(figsize=(12, 6))
avg_price = df.groupby('neighbourhood_group')['price'].mean().sort_values()
sns.barplot(x=avg_price.index, y=avg_price.values, palette="coolwarm")

plt.title('Average Price per Neighbourhood Group', fontsize=18, fontweight='bold')
plt.xlabel('Neighbourhood Group', fontsize=14)
plt.ylabel('Average Price ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

##### 1. Why did you pick the specific chart?

This bar chart clearly compares average prices across neighborhood groups for quick pricing insights.

##### 2. What is/are the insight(s) found from the chart?

Manhattan has the highest average listing price, while Bronx has the lowest.
